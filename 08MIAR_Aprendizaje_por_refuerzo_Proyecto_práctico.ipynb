## **PARTE 3**. Desarrollo y preguntas

#### Importar librerías

````python
from __future__ import division

from PIL import Image
import numpy as np
import gym

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Flatten, Convolution2D, Permute, BatchNormalization, Dropout
from tensorflow.keras import regularizers, layers, initializers, optimizers
from tensorflow.keras.optimizers import Adam
import tensorflow.keras.backend as K
from tensorflow.keras.callbacks import LearningRateScheduler

from rl.agents.dqn import DQNAgent
from rl.policy import LinearAnnealedPolicy, BoltzmannQPolicy, EpsGreedyQPolicy
from tensorflow.keras.optimizers.schedules import ExponentialDecay
from rl.memory import SequentialMemory
from rl.core import Processor
from rl.callbacks import FileLogger, ModelIntervalCheckpoint
````

#### Configuración base

````python
INPUT_SHAPE = (84, 84)

#Se aunmenta el número de fotogramas para intentandar predecir mejor como se moverán las naves
WINDOW_LENGTH = 16

env_name = 'SpaceInvaders-v0'
env = gym.make(env_name)

np.random.seed(123)
env.seed(123)
nb_actions = env.action_space.n
````

````python
class AtariProcessor(Processor):
    def process_observation(self, observation):
        assert observation.ndim == 3  # (height, width, channel)
        img = Image.fromarray(observation)
        img = img.resize(INPUT_SHAPE).convert('L')
        processed_observation = np.array(img)
        assert processed_observation.shape == INPUT_SHAPE
        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        processed_batch = batch.astype('float32') / 255.
        return processed_batch

    def process_reward(self, reward):
        return np.clip(reward, -1., 1.)
````

````python
print("Numero de acciones disponibles: " + str(nb_actions))
````

````python
print("Formato de las observaciones:")
env.observation_space
````

````python
 #Definir una función para actualizar el learning rate
def lr_schedule(epoch):
    initial_lr = 0.00021  # Learning rate inicial
    factor = 0.5  # Factor de aumento del learning rate
    num_epochs_to_increase = 5  # Número de épocas para aumentar el learning rate
    new_lr = initial_lr * (factor ** (epoch // num_epochs_to_increase))
    return new_lr

# Crear el callback del LearningRateScheduler
lr_callback = LearningRateScheduler(lr_schedule)
````

1. Implementación de la red neuronal

````python
convnet = Sequential()
_input_shape = (WINDOW_LENGTH,) + INPUT_SHAPE
# 2D Convolution layer 1
convnet.add(layers.Conv2D(32, (8, 8),padding='same', activation='relu', input_shape=_input_shape)) 
convnet.add(layers.MaxPooling2D((2, 2))) 
# 2D Convolution layer 2
convnet.add(layers.Conv2D(64, (4, 4),padding='same', activation='relu'))
convnet.add(layers.MaxPooling2D((2, 2)))
# 2D Convolution layer 3
convnet.add(layers.Conv2D(64, (3, 3),padding='same', activation='relu'))
#convnet.add(layers.MaxPooling2D((2, 2)))
convnet.add(Dropout(0.2))
# Primer (y único) set de capas FC => RELU
convnet.add(layers.Flatten()) 
convnet.add(layers.Dense(512, activation='relu'))
#Add Output Layer
convnet.add(layers.Dense(nb_actions, activation='linear'))
print(convnet.summary())
````

2. Implementación de la solución DQN

````python
memory = SequentialMemory(limit=100000, window_length=WINDOW_LENGTH)
processor = AtariProcessor()
````

````python
policy = LinearAnnealedPolicy(EpsGreedyQPolicy(), attr='eps',
                              value_max=1., value_min=.1, value_test=.05,
                              nb_steps=100000)

policy = BoltzmannQPolicy(tau=0.2)
````

````python
dqn = DQNAgent(model=convnet, nb_actions=nb_actions, policy=policy,
               memory=memory, processor=processor,
               nb_steps_warmup=50000, gamma=.20,
               target_model_update=1000,
               train_interval=512)
dqn.compile(Adam(learning_rate=.000025), metrics=['mae'])
````

````python
# Training part
weights_filename = 'dqn_{}_weights.h5f'.format(env_name)
checkpoint_weights_filename = 'dqn_' + env_name + '_weights_{step}.h5f'
log_filename = 'dqn_{}_log.json'.format(env_name)
callbacks = [ModelIntervalCheckpoint(checkpoint_weights_filename, interval=250000)]
callbacks += [FileLogger(log_filename, interval=100)]

dqn.fit(env, callbacks=callbacks, nb_steps=1750000, log_interval=10000, visualize=False)

dqn.save_weights(weights_filename, overwrite=True)
````

````python
# Testing part to calculate the mean reward
weights_filename = 'dqn_{}_weights.h5f'.format(env_name)
dqn.load_weights(weights_filename)
dqn.test(env, nb_episodes=10, visualize=True)
````

3. Justificación de los parámetros seleccionados y de los resultados obtenidos

En este trabajo, hemos explorado diferentes estrategias de resolución del problema y hemos analizado sus ventajas e inconvenientes. En nuestra aproximación, nos hemos inspirado en los conceptos vistos en clase para abordar el desafío. Para comprender mejor las decisiones tomadas al parametrizar nuestro modelo, es importante comprender cómo se desarrolla el juego.

Las mayores recompensas se obtienen al eliminar la mayor cantidad posible de enemigos, los cuales se mueven constantemente y disparan. Consideramos que utilizar solo una ventana de 4 imágenes no sería suficiente para captar información sobre la dirección de movimiento de los enemigos, si están descendiendo o si están disparando. Por lo tanto, decidimos utilizar una línea temporal más larga para tener una mejor perspectiva y mejorar nuestras predicciones.

En cuanto a la selección de la política de acción, optamos por la BoltzmannQPolicy con un valor de tau de 0.2. Un valor más alto de tau haría que las acciones se seleccionen de manera más aleatoria, mientras que un valor más bajo daría más prioridad a las acciones prometedoras. No queríamos tomar decisiones demasiado aleatorias y preferimos dar más importancia a las acciones que se consideran más prometedoras. Dado que no estamos tratando con un entorno que tiene una gran cantidad de acciones posibles, no necesitamos explorar de manera exhaustiva.

Otro factor que ajustamos para buscar una mayor puntuación fue la gamma, establecida en 0.20. Un valor bajo hace que las recompensas inmediatas tengan más peso que las futuras. Dado que nuestro objetivo era relativamente bajo, consideramos que sería beneficioso obtener múltiples recompensas altas rápidamente, como eliminar a varios enemigos antes de realizar cualquier movimiento.

Si bien en la etapa de prueba logramos superar el objetivo en una ocasión, en los demás casos no lo conseguimos. El promedio de puntuación se situó por debajo de nuestras expectativas. Quizás podríamos plantear una estrategia más conservadora que permita obtener pequeñas recompensas de forma constante a lo largo del tiempo
